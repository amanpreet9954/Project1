## Program 01:

Write a Python program which iterates through the integers from 1 to 100. 

   - If the number is divisible by 3, it should print "Fizz".

## Program 02:
Write a program to compute the frequency of the words from the input. 
The output should output after sorting the key alphanumerically.
Suppose the following input is supplied to the program:

#### Input:
New to Python or choosing between Python 2 and Python 3? Read Python 2 or Python 3.

#### Output:
2:2<br>
3.:1<br>
3?:1<br>
New:1<br>
Python:5<br>
Read:1<br>
and:1<br>
between:1<br>
choosing:1<br>
or:2<br>
to:1<br>

```python
def compute_word_frequency(text):
    # Split the input text into words based on spaces
    words = text.split()
    
    # Dictionary to store word frequencies
    word_freq = {}
    
    # Calculate word frequencies
    for word in words:
        if word in word_freq:
            word_freq[word] += 1
        else:
            word_freq[word] = 1
            
    # Sort the dictionary by keys (words) alphanumerically
    sorted_word_freq = dict(sorted(word_freq.items()))
    
    return sorted_word_freq

if __name__ == "__main__":
    input_text = "New to Python or choosing between Python 2 and Python 3? Read Python 2 or Python 3."
    word_frequencies = compute_word_frequency(input_text)
    
    # Print the word frequencies
    for word, freq in word_frequencies.items():
        print(f"{word}:{freq}")
```

## Program 03:

Write a program which takes a date of birth of a person and then calculates the age.
(Program must handle edge case scenarios with appropriate error message)

Assumption: User input date of birth should be in DD/MM/YYYY format.

```python
from datetime import datetime

def calculate_age(dob):
    try:
        # Parse the date of birth
        birth_date = datetime.strptime(dob, "%d/%m/%Y")
        
        # Get today's date
        today = datetime.today()
        
        # Calculate age
        age = today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))
        
        return age
    
    except ValueError:
        return "Invalid date format. Please enter the date in DD/MM/YYYY format and with proper values."

if __name__ == "__main__":
    dob = input("Enter your date of birth (DD/MM/YYYY): ")
    age = calculate_age(dob)
    
    if isinstance(age, int):
        print(f"Your age is: {age}")
    else:
        print(age)
```

## Program 04:

Write a program which prints n number of Fibonacci sequence.

Hint: Use recursion method for implementation.

```python

def recur_fibo(n):
    if n <= 1:
        return n
    else:
        return recur_fibo(n-1) + recur_fibo(n-2)

nterms = 11

# check if the number of terms is valid
if nterms <= 0:
    print("Please enter a positive integer")
else:
    print("Fibonacci sequence:")
    for i in range(nterms):
        print(recur_fibo(i))
```

## Program 05:

Write a program to have a count of continuous occurences (more than 1) of a character in front of each character in the string.

#####  Input:
message = "hello"

##### Output:
"hel2o"

#####  Input:
message = "aabbbcddaaa"

##### Output:
"a2b3cd2a3"

```python
def count_continuous_occurrences(msg):
    if not msg:
        return ""

    result = []
    count = 1

    for i in range(1, len(msg)):
        if msg[i] == msg[i - 1]:
            count += 1
        else:
            result.append(msg[i - 1])
            if count > 1:
                result.append(str(count))
            count = 1

    result.append(msg[-1])
    if count > 1:
        result.append(str(count))

    return ''.join(result)

if __name__ == "__main__":
    # Test cases
    msg1 = "hello"
    msg2 = "aabbbcddaaa"

    print(count_continuous_occurrences(msg1))  # Output: "hel2o"
    print(count_continuous_occurrences(msg2))  # Output: "a2b3cd2a3"
```

## Program 06:

Given an array of integers 'nums' and an integer 'target', return indices of the first pair of 2 elements such that they add up to 'target'.

You may assume that each input may have multiple solutions, and same indexed element cannot be used twice.

You can return the answer (indexes) in any order
    
##### Input: nums = [2, 3, 7, 11, 6, 15], Target = 9
##### Output: [0, 2]
    
##### Input: nums = [3, 2, 4, 6, 4], Target = 6
##### Output: [1, 2]

```python
def two_sum(nums, target):
    
    num_to_index = {}
    
    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_to_index:
            return [num_to_index[complement], i]
        num_to_index[num] = i
    
    return []


nums1 = [2, 3, 7, 11, 6, 15]
target1 = 9
print(two_sum(nums1, target1))  

nums2 = [3, 2, 4, 6, 4]
target2 = 6
print(two_sum(nums2, target2))  
```

## Program 07:

Write a program to mask the file name of the S3 Image URI. 

You have been given a json response below and you have to mask the image file name except the first 8 characters and the last 8 characters of the file name (excluding file extension). You can use a static number for masking: 5 asterisks.

If filename is less than 16 characters, don't mask the file name.

Note: Original input response should not be changed.

##### Input:

{<br>
  'Employee ID': '1050',<br>
  'Image': 's3://morningstar/employee-id/data/valid/123456-DRIVING-LICENSE-IND.jpg',<br>
  'id_type': 'Drivers License',<br>
  'environment': 'Development'<br>
}<br>

##### Output:

Original Response:

{<br>
  'Employee ID': '1050',<br>
  'Image': 's3://morningstar/employee-id/data/valid/123456-DRIVING-LICENSE-IND.jpg',<br>
  'id_type': 'Drivers License',<br>
  'environment': 'Development'<br>
}<br>

Masked Response:

{<br>
  'Employee ID': '1050',<br>
  'Image': 's3://morningstar/employee-id/data/valid/123456-D*****ENSE-IND.jpg',<br>
  'id_type': 'Drivers License',<br>
  'environment': 'Development'<br>
}<br>

```python
import json
import re

def mask_s3_image_filename(data):
    
    masked_data = json.loads(json.dumps(data))
    image_uri = masked_data.get('Image', '')
    
    # Find the filename in the URI
    match = re.search(r'([^/]+)(\.\w+)$', image_uri)
    
    if match:
        filename = match.group(1)
        extension = match.group(2)
        
        if len(filename) > 16:
            
            masked_filename = filename[:8] + '*****' + filename[-8:]
            masked_image_uri = image_uri.replace(filename + extension, masked_filename + extension)
            
            masked_data['Image'] = masked_image_uri
    
    return masked_data

# Input data
input_data = {
    'Employee ID': '1050',
    'Image': 's3://morningstar/employee-id/data/valid/123456-DRIVING-LICENSE-IND.jpg',
    'id_type': 'Drivers License',
    'environment': 'Development'
}

# Generate masked response
masked_response = mask_s3_image_filename(input_data)

# Print original and masked responses
print("Original Response:\n")
print(json.dumps(input_data, indent=4))

print("\nMasked Response:\n")
print(json.dumps(masked_response, indent=4))
```

```python
import pandas as pd
import numpy as np

# Load the data from the specified path
data = pd.read_csv('/Users/amanpreet/Downloads/Morningstar_CRQ_Python_test/data.csv')

# Convert CURBALANCE to numeric, handling non-numeric gracefully
data['CURBALANCE'] = pd.to_numeric(data['CURBALANCE'], errors='coerce')

# Extract the spread from GROSSRATEFORMULA
data['Spread'] = data['GROSSRATEFORMULA'].str.extract(r'\+\s*(\d+\.\d+|\d+)').astype(float)

# Calculate the CURBALANCE of loans domiciled outside the US
data['Outside_US_CURBALANCE'] = data['CURBALANCE'] * (data['COUNTRY'] != 'United States').astype(int)

# Define a function to determine the lowest rating
def get_lowest_rating(row):
    ratings = {
        'NR': 0, # NR means No rating so this would be ignored and rest other ratings will be considered
        'AAA': 1, 'AA+': 2, 'AA': 3, 'AA-': 4, 'A+': 5, 'A': 6, 'A-': 7,
        'BBB+': 8, 'BBB': 9, 'BBB-': 10, 'BB+': 11, 'BB': 12, 'BB-': 13,
        'B+': 14, 'B': 15, 'B-': 16, 'CCC+': 17, 'CCC': 18, 'CCC-': 19,
        'CC': 20, 'C': 21, 'D': 22,
        'Aaa': 1, 'Aa1': 2, 'Aa2': 3, 'Aa3': 4, 'A1': 5, 'A2': 6, 'A3': 7,
        'Baa1': 8, 'Baa2': 9, 'Baa3': 10, 'Ba1': 11, 'Ba2': 12, 'Ba3': 13,
        'B1': 14, 'B2': 15, 'B3': 16, 'Caa1': 17, 'Caa2': 18, 'Caa3': 19,
        'Ca': 20
    }
    ratings_list = [
        ratings.get(row['RATING_1'], 23),
        ratings.get(row['RATING_2'], 23),
        ratings.get(row['RATING_3'], 23)
    ]
    return max(ratings_list) # taking max because we are going on increasing order of ratings to be lowest

# Apply the function to each row
data['Lowest_Rating'] = data.apply(get_lowest_rating, axis=1)

print(data[['Deal', 'Loan', 'RATING_1', 'RATING_2', 'RATING_3', 'Lowest_Rating']])

# Filter out defaulted loans
non_defaulted_data = data[data['DEFAULT_FLAG'] != 'Y']

# Group by 'Deal' and 'YearMonth'
grouped = non_defaulted_data.groupby(['Deal', 'YearMonth'])

# Aggregate the required metrics
result = grouped.agg(
    Total_Loan_Count=('Loan', 'size'),  # Counting loans explicitly
    Total_CURBALANCE=('CURBALANCE', 'sum'),
    Outside_US_CURBALANCE=('Outside_US_CURBALANCE', 'sum'),
    Weighted_Avg_Spread=('CURBALANCE', lambda x: np.average(non_defaulted_data.loc[x.index, 'Spread'], weights=x)),
    Loans_Below_CCCPlus=('Lowest_Rating', lambda x: (x > 17).sum())  # Count loans with rating below CCC+
).reset_index()

# Calculate the percentage of CURBALANCE outside the US
result['Percent_Outside_US'] = (result['Outside_US_CURBALANCE'] / result['Total_CURBALANCE']) * 100

# Calculate the percentage of loans below CCC+
result['Percent_Below_CCCPlus'] = (result['Loans_Below_CCCPlus'] / result['Total_Loan_Count']) * 100

# Reorder columns for display
result = result[['Deal', 'YearMonth', 'Total_Loan_Count', 'Total_CURBALANCE',
                 'Weighted_Avg_Spread', 'Percent_Outside_US', 'Percent_Below_CCCPlus']]

# Print the final result
print(result)

result.to_csv('/Users/amanpreet/Downloads/Morningstar_CRQ_Python_test/data_results.csv')
```
